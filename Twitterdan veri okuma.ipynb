```python
def detect_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
df2=tdf[["text"]]
```

```python
df2.head()
```

```python
df2["sentiment"]=df2["text"].apply(detect_sentiment)
```

```python
df2.head()
```

```python
df2.sentiment.value_counts()
```

```python
df2["text"]=df2["text"].str.lower()
```

```python
df2["text"]=df2["text"].str.replace("[^\w\s]","")
```

```python
df2["text"]=df2["text"].str.replace("\d+","")
```

```python
df2["text"]=df2["text"].str.replace("\n","").replace("\r","")
```

```python
from wordcloud import STOPWORDS
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
def wc(data,bgcolor):
    plt.figure(figsize=(10,10))
    mask=np.array(Image.open("togg.jpg"))
    wc=WordCloud(background_color=bgcolor,stopwords=STOPWORDS,mask=mask)
    wc.generate(' '.join(data))
    plt.imshow(wc)
    plt.axis("off")
```

```python
wc(df2.text,'white')
```

```python
#1=pozitive
#2=negative
#0=neutral
def sentiment2(sent):
    if sent <-0.2:
        return 2
    elif sent>-0.2:
        return 1
    else: 
        return 0
    
```

```python
df2["sent"]=df2["sentiment"].apply(sentiment2)
```

```python
import seaborn as sns
```

```python
sns.countplot(df2.sent)
```

```python
x=df2.text
y=df2.text
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
vect=CountVectorizer(stop_words="english")
```

```python
x=vect.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.15)
```

```python
mb=MultinomialNB()
```

```python
tahmin=mb.fit(x_train,y_train).predict(x_test)
```

```python
accuracy_score(tahmin,y_test) # verimiz 100 satirlik
```
